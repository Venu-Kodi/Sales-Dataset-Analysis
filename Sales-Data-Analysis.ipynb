Loading the dataset and Merging the 12 Months sales data into a single file

```python
df = pd.read_csv("Sales_April_2019.csv")
df.head()
```

```python
path = './Downloads/Sales_Data'
extension = '.csv'
```

```python
files = [file for file in os.listdir(path) if file.endswith(extension)]
```

```python
dfs = []
for file in files:
    df = pd.read_csv(os.path.join(path, file))
    dfs.append(df)
```

```python
df = pd.concat(dfs, ignore_index=True)
```

```python
df.head()
```

```python
df.tail(5)
```

```python
#Clean the data
# Below found the nan values by creating new df1 
```

```python
df1 = df[df.isna().any(axis=1)]
df1.head()
```

```python
df['Month'] = df['Order Date'].str[0:2]
```

```python
#Dropping the nan
df = df.dropna(how='all')
df.head()
```

```python
#changing 4/ to 04
df['Month'] = df['Month'].str.replace('4/','04')
df['Month'] = df['Month'].str.replace('5/','05')
print(df)
```

```python
df = df[df['Order Date'].str[0:2] != 'Or']
```

```python
df.head()
```

```python
#Augment data with additional columns
```

```python
#Adding Month column
```

```python
df['Month'] = df['Month'].astype('int32')
df.head()
```

```python
#Adding a sales column
```

```python
df.dtypes
```

```python
#convert columns to the correct type
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
df.head()
```

```python
df.dtypes
```

```python
df['Sales'] = df['Quantity Ordered']*df['Price Each']
df.head()
```

What was the best month for sales? How much was earned that month?

```python
results = df.groupby('Month').sum()
```

```python
df.groupby('Month').sum()['Sales']
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month')
plt.show()
```

What products are most often sold together?

```python
df['Order Date'] = df['Order Date'].replace('%m/%d/%Y %H:%M', '4/19/2019 8:46')
df['Order Date']
```

```python
df.head()
```

```python
df1 = df[df['Order ID'].duplicated(keep=False)]
df1.head(10)
```

```python
df1['Grouped'] = df1.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df1.head()
```

```python
df1 = df1[['Order ID', 'Grouped']].drop_duplicates()
df1.head()
```

```python
from itertools import combinations
from collections import counter

count = counter()

for row in df1['Grouped']:
    row_list = row.split(',')
    count.update(counter(combinations(row_list,2)))

print(count)
```

```python
count.most_common(10)
```
